### Importaci√≥n de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. An√°lisis de facturaci√≥n



```python
for df in [tienda, tienda2, tienda3, tienda4]:
  df["Facturacion"] = df["Precio"] + df["Costo de env√≠o"]
```

```python
print("Facturacion total por tienda:")
print(f"Tienda 1: ${tienda['Facturacion'].sum():,.0f}")
print(f"Tienda 2: ${tienda2['Facturacion'].sum():,.0f}")
print(f"Tienda 3: ${tienda3['Facturacion'].sum():,.0f}")
print(f"Tienda 4: ${tienda4['Facturacion'].sum():,.0f}")
```

```python
import matplotlib.pyplot as plt

facturaciones = {
    "Tienda 1": tienda["Facturacion"].sum(),
    "Tienda 2": tienda2["Facturacion"].sum(),
    "Tienda 3": tienda3["Facturacion"].sum(),
    "Tienda 4": tienda4["Facturacion"].sum()
}

plt.bar(facturaciones.keys(), facturaciones.values(), color="skyblue", edgecolor="black")
plt.title("üí∞ Facturaci√≥n Total por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Monto total ($)")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Ventas por categor√≠a

```python
print("Ventas por categor√≠a - Tienda 1")
categoria_t1 = tienda.groupby("Categor√≠a del Producto")["Facturacion"].sum().sort_values(ascending=False)
display(categoria_t1)
```

```python
print("Ventas por categor√≠a - Tienda 2")
categoria_t2 = tienda2.groupby("Categor√≠a del Producto")["Facturacion"].sum().sort_values(ascending=False)
display(categoria_t2)
```

```python
print("Ventas por categor√≠a - Tienda 3")
categoria_t3 = tienda3.groupby("Categor√≠a del Producto")["Facturacion"].sum().sort_values(ascending=False)
display(categoria_t3)
```

```python
print("Ventas por categor√≠a - Tienda 4")
categoria_t4 = tienda4.groupby("Categor√≠a del Producto")["Facturacion"].sum().sort_values(ascending=False)
display(categoria_t4)
```

```python
import matplotlib.pyplot as plt

tiendas = [
    ("Tienda 1", categoria_t1),
    ("Tienda 2", categoria_t2),
    ("Tienda 3", categoria_t3),
    ("Tienda 4", categoria_t4)
]

for nombre, df in tiendas:
    plt.figure(figsize=(8,4))
    df.plot(kind="bar", color="lightcoral", edgecolor="black")
    plt.title(f"üí∞ Facturaci√≥n por Categor√≠a - {nombre}")
    plt.ylabel("Monto total ($)")
    plt.xlabel("Categor√≠a del Producto")
    plt.xticks(rotation=45, ha="right")
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()
```

# 3. Calificaci√≥n promedio de la tienda


```python
promedios = {
    "Tienda 1": tienda["Calificaci√≥n"].mean(),
    "Tienda 2": tienda2["Calificaci√≥n"].mean(),
    "Tienda 3": tienda3["Calificaci√≥n"].mean(),
    "Tienda 4": tienda4["Calificaci√≥n"].mean()
}

for nombre, valor in promedios.items():
    print(f"{nombre}: {valor:.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
bars = plt.bar(promedios.keys(), promedios.values(), color="gold", edgecolor="black")
plt.title("Calificaci√≥n Promedio por Tienda", fontsize=14, fontweight="bold")
plt.xlabel("Tienda")
plt.ylabel("Calificaci√≥n promedio")
plt.ylim(0, 5)
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar el valor sobre cada barra
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f"{bar.get_height():.2f}", ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
